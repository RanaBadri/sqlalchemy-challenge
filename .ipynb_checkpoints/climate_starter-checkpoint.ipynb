# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
sqlalchemy.__version__  
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
Base = automap_base()
```

```python
# reflect an existing database into a new model
Base.prepare(engine, reflect=True)
# reflect the tables
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
measurement = Base.classes.measurement
station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
import datetime
```

```python
#obtained the format for dates
sample_dates = session.query(measurement.date).limit(5).all()
for result in sample_dates:
    print(result)
```

```python
#identified the last date of data provided
max_date = session.query(func.max(func.strftime("%Y-%m-%d", measurement.date))).limit(5).all()
max_date[0][0]
```

```python
#obtained perciptation date for the last year available in the data provided
precip_data = session.query(func.strftime("%Y-%m-%d", measurement.date), measurement.prcp).\
    filter(func.strftime("%Y-%m-%d", measurement.date) >= datetime.date(2016, 8,23)).all()
precip_data
```

```python
#Select only the date and prcp values
#Load the query results into a Pandas DataFrame and set the index to the date column.

precp_df = pd.DataFrame(precip_data, columns = ['date', 'precipitation'])

#index
precp_df.set_index('date', inplace = True)

#look at dataframe
precp_df.head()
```

```python
#Sort the DataFrame values by date.
precp_df = precp_df.sort_values(by = 'date')
precp_df.head()
```

```python
#Plot the results using the DataFrame plot method.
fig, ax = plt.subplots(figsize = (15, 7))
precp_df.plot(ax = ax, x_compat = True)

ax.set_ylabel('Precipitation (in.)')

plt.savefig("Images/precip.png")

plt.tight_layout()
plt.show()
```

```python
#Station Analysis
```

```python
#Design a query to calculate the total number of stations.
station_count = session.query(station.id).distinct().count()
station_count
```

```python
#Design a query to find the most active stations.
#List the stations and observation counts in descending order.

activ_stations = session.query(station.station, func.count(measurement.id)).select_from(measurement).\
    join(station, measurement.station == station.station).group_by(station.station).\
    order_by(func.count(measurement.id).desc()).all()

for result in activ_stations:
    print(f"Station: {result[0]}\tCount: {result[1]}")
```

```python
#Which station has the highest number of observations?
most_active_station = 'USC00519281'
```

```python
#Design a query to retrieve the last 12 months of temperature observation data (TOBS).
tobs_lastyear = session.query(func.strftime("%Y-%m-%d", measurement.date), measurement.tobs).\
    filter(func.strftime("%Y-%m-%d", measurement.date) >= datetime.date(2016, 8,23)).all()

tobs_lastyear_df = pd.DataFrame(tobs_lastyear, columns = ['date', 'temperature'])
tobs_lastyear_df.head()
```

```python
#Plot the results as a histogram with bins=12.
fig, ax = plt.subplots()
tobs_lastyear_df.plot.hist(bins = 12, ax = ax)

plt.savefig("Images/temphist.png")
plt.tight_layout()
plt.show()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
# df.describe()

summary_temps = session.query(func.min(measurement.tobs), func.max(measurement.tobs), func.avg(measurement.tobs)).\
   filter(measurement.station == most_active_station).all()
print(f"Lowest Temperature: {summary_temps[0][0]} Fahrenheit")
print(f"Highest Temperature: {summary_temps[0][1]} Fahrenheit")
print(f"Average Temperature: {round(summary_temps[0][2], 2)} Fahrenheit")
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(measurement.tobs), func.avg(measurement.tobs), func.max(measurement.tobs)).\
        filter(measurement.date >= start_date).filter(measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.

def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(measurement.tobs), func.avg(measurement.tobs), func.max(measurement.tobs)).\
        filter(measurement.date >= start_date).filter(measurement.date <= end_date).all()

# function usage example
print(calc_temps('2016-03-25', '2016-03-31'))
```

```python
# Calculate the total amount of rainfall per weather station for your trip dates using the previous year's matching dates.
# Sort this in descending order by precipitation amount and list the station, name, latitude, longitude, and elevation

Trip_rainfall = session.query(station.station, station.name, station.latitude, station.longitude,
                               station.elevation, func.avg(measurement.prcp)).\
    filter(measurement.station == station.station).\
    filter(func.strftime("%Y-%m-%d", measurement.date) >= datetime.date(2015, 3, 25)).group_by(station.station).\
    order_by(func.avg(measurement.prcp).desc()).all()

Trip_rainfall
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(measurement.tobs), func.avg(measurement.tobs), func.max(measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", measurement.date) == date).all()
    
daily_normals("01-01")
```

```python
# calculate the daily normals for your trip
my_trip_dates = ["03-25", "03-26", "03-27", "03-28", "03-29", "03-30", "03-31"]

trip_weather = []

for my_trip_date in my_trip_dates:
    day = {}
    day['Date'] = f"2016-{my_trip_date}"
    day['Min'] = daily_weather[0][0]
    day['Avg'] = daily_weather[0][1]
    day['Max'] = daily_weather[0][2]
    daily_weather.append(day)
    
trip_weather
```
